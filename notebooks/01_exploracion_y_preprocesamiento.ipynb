# Histograma de `SalePrice`
Tomaremos la cantidad de bins basandomos en la regla de Freedman-Diaconis
Ya tenemos Q3 que es 214000 y Q1 que es 129975
Max - Min = 755000 - 34900 = 720100

```python
# Freedman-Diaconis rule
q3 = 214000
q1 = 129975
iqr = q3 - q1
anchoBin = 2 * iqr / (len(dataframe) ** (1/3))
binsNumber = int((720100) / anchoBin)
print(f"Ancho del bin: {anchoBin}, Numero de bins: {binsNumber}")
dataframe["SalePrice"].hist(bins=binsNumber)
```

Sin embargo podemos ver que hay algunos outliers, es decir, casas con un área habitable grande pero con un precio de venta relativamente bajo.

## Manejo de datos faltantes

```python
conteo_nulos = dataframe.isnull().sum()
print(conteo_nulos)
```

Con esto podemos ver que hay dos columnas con datos nulos, `LotFrontage` y `Alley` con 259 y 1369 valores nulos respectivamente.

Para esto mi estrategia sera la siguiente:
- Para `Alley`, segun la descripcion de los datos existe la categoria NA que significa "No Alley Access", por lo que reemplazare los valores nulos con esta categoria.
- Para `LotFrontage`, que segun la descripcion de los datos es la longitud lineal del frente de la calle adyacente a la propiedad, reemplazare los valores nulos con un dato segun la vecindad (Neighborhood) a la que pertenece la propiedad. Porque normalmente las propiedades en la misma vecindad tienen caracteristicas similares.

```python
# Rellenar datos nulos de Alley con NA
dataframe['Alley'] = dataframe['Alley'].fillna('NA')

# Rellenar datos nulos de LotFrontage con la mediana por vecindad
dataframe['LotFrontage'] = dataframe.groupby('Neighborhood')['LotFrontage'].transform(
    lambda x: x.fillna(x.median())
)

# Revisar que ya no hay datos nulos
conteo_nulos = dataframe.isnull().sum()
print(conteo_nulos)
```

# Normalizacion de datos

Convertiremos los datos categoricos en numericos usando One-Hot Encoding con la funcion `get_dummies` de pandas.

```python
# Columans con datos binarios para eliminar una de las dos categorias y evitar la multicolinealidad
binary_cols = ['Street', 'CentralAir']
dataframe = pd.get_dummies(dataframe, columns=binary_cols, drop_first=True)

# Lo mismo para las columnas categoricas pero con mas de dos categorias y aca si mantener todas las columnas
categorical_cols = dataframe.select_dtypes(include=['object']).columns
dataframe = pd.get_dummies(dataframe, columns=categorical_cols, drop_first=False) 

print("Columnas después de One-Hot Encoding:")
print(dataframe.columns)
```
